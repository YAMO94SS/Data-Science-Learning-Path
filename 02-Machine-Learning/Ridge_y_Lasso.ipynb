```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import os # Importamos la librería para el manejo de rutas
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# 1. Configuramos la ruta exacta a tu nueva carpeta de trabajo
os.chdir(r'C:\Users\Yamil\OneDrive\Escritorio\Escritorio\Yamil\EBAC (CIENCIA DE DATOS)\Bloque 4 Aprendizaje de máquina supervisado\MODULO 2 Regresión regularizada Ridge y Lasso\Práctica')

# 2. Cargamos el dataset y separamos características (X) y variable objetivo (y)
df = pd.read_excel('FuelConsumptionCo2.xlsx')

X = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB']]
y = df['CO2EMISSIONS']

# División 80% entrenamiento y 20% prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Modelo 1: Regresión Lineal Múltiple (Ejercicio 3)
linreg = LinearRegression()
linreg.fit(X_train, y_train)
y_pred_lin = linreg.predict(X_test)

print("--- 1. Regresión Lineal Múltiple ---")
print(f"R2: {r2_score(y_test, y_pred_lin):.4f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_lin):.4f}")
print(f"MSE: {mean_squared_error(y_test, y_pred_lin):.4f}\n")

# Modelo 2: Regresión Ridge con Alpha óptimo (Ejercicio 4)
# Creamos un rango de alphas para probar
alpha_range = 10.**np.arange(-3, 5)
# Usamos un pipeline para escalar los datos (reemplazo de normalize=True)
pipe_ridge = make_pipeline(StandardScaler(), RidgeCV(alphas=alpha_range, scoring='neg_mean_squared_error'))
pipe_ridge.fit(X_train, y_train)
y_pred_ridge = pipe_ridge.predict(X_test)
alpha_optimo_ridge = pipe_ridge.named_steps['ridgecv'].alpha_

print(f"--- 2. Regresión Ridge (Alpha Óptimo: {alpha_optimo_ridge}) ---")
print(f"R2: {r2_score(y_test, y_pred_ridge):.4f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_ridge):.4f}")
print(f"MSE: {mean_squared_error(y_test, y_pred_ridge):.4f}\n")

# Modelo 3: Regresión Lasso con Alpha óptimo (Ejercicio 5)
pipe_lasso = make_pipeline(StandardScaler(), LassoCV(cv=5, random_state=42))
pipe_lasso.fit(X_train, y_train)
y_pred_lasso = pipe_lasso.predict(X_test)
alpha_optimo_lasso = pipe_lasso.named_steps['lassocv'].alpha_

print(f"--- 3. Regresión Lasso (Alpha Óptimo: {alpha_optimo_lasso:.4f}) ---")
print(f"R2: {r2_score(y_test, y_pred_lasso):.4f}")
print(f"MAE: {mean_absolute_error(y_test, y_pred_lasso):.4f}")
print(f"MSE: {mean_squared_error(y_test, y_pred_lasso):.4f}")
```
