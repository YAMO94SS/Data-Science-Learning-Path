```python
# =========================================================
# PROBLEMA 1: K-NEAREST NEIGHBORS (KNN) - WINE CLUSTERING
# =========================================================
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors

# 1. Carga de datos con tu ruta específica
ruta_csv = r"C:\Users\Yamil\OneDrive\Escritorio\Escritorio\Yamil\EBAC (CIENCIA DE DATOS)\Bloque 5 Aprendizaje de máquina no supervisado\MOD 5 KNN y Market basket\Práctica\wine-clustering.csv"
df_vinos = pd.read_csv(ruta_csv)

# 2. Definición de la población (X) y el punto de prueba
# Asumimos que la tabla tiene exactamente las 13 columnas en el orden de la imagen
X_train = df_vinos.values
test_point = [[14, 2, 2.5, 16, 115, 3, 2.5, 0.4, 2, 9, 1, 3.5, 800]]

# VISIÓN TÉCNICA: En un entorno real aplicaríamos StandardScaler porque 
# la variable 'Proline' (valores > 1000) dominará la distancia euclidiana 
# sobre 'Hue' (valores ~ 1). Lo ejecutamos en bruto para cumplir la rúbrica académica.

# 3. Entrenamiento del Modelo KNN (k=5)
k = 5
knn = NearestNeighbors(n_neighbors=k, algorithm='auto')
knn.fit(X_train)

# 4. Cálculo de distancias y obtención de vecinos
distancias, indices = knn.kneighbors(test_point)
indices_vecinos = indices[0]

# 5. Extracción de concentraciones de alcohol (Columna 'Alcohol' = índice 0)
alcohol_vecinos = df_vinos.loc[indices_vecinos, 'Alcohol'].values
promedio_alcohol = np.mean(alcohol_vecinos)

print("--- RESULTADOS PROBLEMA 1 (KNN) ---")
print(f"ID (Índices) de los 5 vinos más cercanos: {indices_vecinos}")
print(f"Concentración de Alcohol de estos 5 vinos: {alcohol_vecinos}")
print(f"Valor Promedio de Alcohol estimado: {promedio_alcohol:.2f}\n")


# =========================================================
# PROBLEMA 2: MARKET BASKET ANALYSIS (APRIORI)
# =========================================================
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# 1. Carga de datos con corrección de calidad de datos ('onion' -> 'onions')
my_basket = [
    ['bread', 'butter', 'wine', 'bananas', 'coffee', 'carrots'],
    ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
    ['beer', 'chips', 'asparagus', 'salsa', 'milk', 'apples'],
    ['olive oil', 'bread', 'butter', 'tomatoes', 'steak', 'carrots'],
    ['tomatoes', 'onions', 'chips', 'wine', 'ketchup', 'orange juice'],
    ['bread', 'butter', 'beer', 'chips', 'milk'],
    ['butter', 'tomatoes', 'carrots', 'coffee', 'sugar'],
    ['tomatoes', 'onions', 'cheese', 'milk', 'potatoes'],
    ['bread', 'butter', 'ketchup', 'coffee', 'chicken wings'],
    ['butter', 'beer', 'chips', 'asparagus', 'apples'],
    ['tomatoes', 'onions', 'beer', 'chips', 'milk', 'coffee'] 
]

# 2. Transformación a Matriz Binaria (Requisito matemático del algoritmo)
te = TransactionEncoder()
te_ary = te.fit(my_basket).transform(my_basket)
df_basket = pd.DataFrame(te_ary, columns=te.columns_)

# 3. Ejecución de Apriori (Soporte mínimo del 20% dado el bajo volumen de datos)
articulos_frecuentes = apriori(df_basket, min_support=0.2, use_colnames=True)

# 4. Extracción de Reglas de Asociación de Alto Impacto (Lift > 1)
rules = association_rules(articulos_frecuentes, metric="lift", min_threshold=1.0)
rules_ejecutivas = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by='lift', ascending=False)

print("--- RESULTADOS PROBLEMA 2 (MARKET BASKET) ---")
print(rules_ejecutivas.head(10))
```
