```python
import sqlite3
import pandas as pd
import os

# 1. Definimos la ruta de trabajo
ruta = r"C:\Users\Yamil\OneDrive\Escritorio\Escritorio\Yamil\EBAC (CIENCIA DE DATOS)\Bloque 7 Análisis de datos con SQL\MOD 8 Consult comun SQL Python\Práctica"
os.chdir(ruta)

# 2. Leemos el CSV con pandas (es la forma más limpia y eficiente para crear la tabla)
df_rh = pd.read_csv("recursos_humanos.csv")

# 3. Creamos la conexión a la base de datos 
conn = sqlite3.connect("RH.db")

# 4. Volcamos el DataFrame a la tabla "Detalle"
# if_exists='replace' asegura que si corres la celda varias veces, no se dupliquen los datos.
df_rh.to_sql('Detalle', conn, if_exists='replace', index=False)

# 5. Comprobamos que la tabla existe y vemos los primeros registros
query_verificacion = "SELECT * FROM Detalle LIMIT 5;"
df_verificacion = pd.read_sql_query(query_verificacion, conn)

# Mostramos el resultado
df_verificacion
```

```python
query_satisfaccion = """
SELECT 
    "left" AS abandona_empresa, 
    AVG(satisfaction_level) AS nivel_satisfaccion_promedio
FROM Detalle
GROUP BY "left";
"""
df_satisfaccion = pd.read_sql_query(query_satisfaccion, conn)
df_satisfaccion
```

```python
query_horas = """
SELECT 
    AVG(average_montly_hours) AS promedio_horas_mensuales
FROM Detalle
WHERE salary IN ('low', 'medium');
"""
df_horas = pd.read_sql_query(query_horas, conn)
df_horas
```

```python
query_fuga_talento = """
SELECT *
FROM Detalle
WHERE promotion_last_5years = 1 AND "left" = 1;
"""
df_fuga_talento = pd.read_sql_query(query_fuga_talento, conn)
df_fuga_talento
```

```python
query_evaluacion = """
SELECT *
FROM Detalle
WHERE last_evaluation >= 0.9;
"""
df_evaluacion = pd.read_sql_query(query_evaluacion, conn)

# Mostramos cuántos empleados cumplen esta condición y los primeros registros
print(f"Total de empleados con evaluación sobresaliente: {len(df_evaluacion)}")
df_evaluacion.head()
```
