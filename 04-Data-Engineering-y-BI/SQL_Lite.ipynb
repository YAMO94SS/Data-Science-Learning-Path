```python
import sqlite3
import pandas as pd
import os

# 1. Definición de la ruta de trabajo local
ruta_trabajo = r"C:\Users\Yamil\OneDrive\Escritorio\Escritorio\Yamil\EBAC (CIENCIA DE DATOS)\Bloque 7 Análisis de datos con SQL\MOD 7 SQLite en Python\Practica"
os.chdir(ruta_trabajo)

# 2. Lectura del dataset y selección de los primeros 5 registros
df_clientes = pd.read_csv("Cientifico de datos M40 Customers.csv")
df_top5 = df_clientes.head(5)

# Conversión de los registros a una lista de tuplas para la inserción en SQLite
registros = [tuple(x) for x in df_top5.to_numpy()]

# 3. Conexión y creación de la Base de Datos (Base.db) mediante Context Manager
with sqlite3.connect("Base.db") as conn:
    cur = conn.cursor()
    
    # 4. Creación de la tabla Clientes definiendo la estructura y tipos de datos
    cur.execute("""
    CREATE TABLE IF NOT EXISTS Clientes (
        CustomerID INTEGER PRIMARY KEY,
        CustomerName TEXT,
        ContactName TEXT,
        Address TEXT,
        City TEXT,
        PostalCode TEXT,
        Country TEXT,
        PurchaseUSD REAL
    );
    """)
    
    # 5. Inserción masiva de los registros utilizando executemany
    cur.executemany("""
    INSERT INTO Clientes (CustomerID, CustomerName, ContactName, Address, City, PostalCode, Country, PurchaseUSD)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?);
    """, registros)
    
    conn.commit()

print("Ejecución exitosa: Base.db creada y registros insertados en la ruta:", os.getcwd())
```

```python
import sqlite3
import os

# Verificamos que seguimos en tu ruta de trabajo por seguridad
print("Generando dump en:", os.getcwd())

# 1. Nos conectamos a la base de datos existente
with sqlite3.connect("Base.db") as conn:
    
    # 2. Abrimos (o creamos) el archivo Operaciones.sql en modo escritura ("w")
    with open("Operaciones.sql", "w", encoding="utf-8") as _file:
        
        # 3. Iteramos sobre el dump de la conexión y escribimos línea por línea
        for line in conn.iterdump():
            _file.write(f"{line}\n")
            
print("Ejecución exitosa: Archivo 'Operaciones.sql' generado correctamente.")
```
