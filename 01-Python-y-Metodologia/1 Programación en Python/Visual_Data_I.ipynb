# Bloque 1: Histograma de Edades (hist)

```python
# Importamos las librerías necesarias
import pandas as pd          # Para manejar datos en DataFrames
import matplotlib.pyplot as plt  # Para crear gráficos

# Cargamos el archivo CSV en un DataFrame de pandas
df = pd.read_csv('titanic.csv')

# Creamos una figura y un eje (axes) donde dibujar el gráfico
fig, ax = plt.subplots()

# --- Creación del histograma ---
# Usamos ax.hist() para generar un histograma de la columna 'Age'
# El resultado se guarda en la variable 'hist' como pide el ejercicio.
# ax.hist() devuelve una tupla con (n, bins, patches), que contiene:
#   - n: las frecuencias de cada bin (array)
#   - bins: los bordes de los bins (array)
#   - patches: los objetos de las barras (para personalización)
# Parámetros importantes:
#   x = df['Age']: los datos a graficar (edades de todos los pasajeros)
#   bins = 10: número de intervalos (barras) del histograma. El test espera exactamente 10 bins.
#   edgecolor = 'black': (opcional) añade un borde negro a las barras para mejorar la visibilidad.
#   alpha = 0.7: (opcional) establece la transparencia de las barras (0 = transparente, 1 = opaco).
hist = ax.hist(df['Age'], bins=10, edgecolor='black', alpha=0.7)

# --- Personalización del gráfico (opcional, no afecta la variable 'hist') ---
ax.set_xlabel('Edad (años)')           # Etiqueta del eje X
ax.set_ylabel('Frecuencia (número de pasajeros)')  # Etiqueta del eje Y
ax.set_title('Histograma de edades de los pasajeros del Titanic')  # Título del gráfico
ax.grid(axis='y', linestyle='--', alpha=0.3)  # Añade una cuadrícula horizontal punteada y semitransparente

# Mostramos el gráfico en pantalla
plt.show()
```

# Bloque 2: Diagrama de Caja (Boxplot) de Edades (box)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')

fig, ax = plt.subplots()

# --- Creación del diagrama de caja ---
# ax.boxplot() dibuja un boxplot a partir de los datos.
# El resultado se guarda en la variable 'box'. Este comando devuelve un diccionario con
# los componentes del boxplot, como:
#   - 'boxes': las cajas (percentiles 25-75)
#   - 'medians': la línea de la mediana
#   - 'whiskers': los bigotes (mínimo y máximo no atípicos)
#   - 'caps': los extremos de los bigotes
#   - 'fliers': los valores atípicos (outliers)
# Parámetros:
#   x = df['Age']: datos de edad de todos los pasajeros.
#   patch_artist = True: (opcional) rellena las cajas con color (por defecto son vacías).
#   notch = False: (opcional) si se quiere una muesca en la mediana para comparar intervalos de confianza.
#   vert = True: (opcional) orientación vertical (True) u horizontal (False).
box = ax.boxplot(df['Age'], patch_artist=True)

# --- Personalización del gráfico ---
ax.set_ylabel('Edad (años)')  # Etiqueta del eje Y
ax.set_title('Diagrama de caja de las edades de los pasajeros del Titanic')  # Título
ax.grid(axis='y', linestyle='--', alpha=0.7)  # Cuadrícula horizontal

plt.show()
```

# Bloque 3: Gráfico de Barras del Conteo por Clase (bar)

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('titanic.csv')

fig, ax = plt.subplots()

# --- Preparación de los datos ---
# Primero, contamos cuántos pasajeros hay en cada clase usando value_counts().
# value_counts() cuenta las ocurrencias de cada valor único en la columna 'Pclass'.
# El resultado es una Serie con los valores únicos como índice y sus frecuencias como valores.
# Luego, sort_index() ordena por el número de clase (1, 2, 3) para que las barras aparezcan en orden lógico.
conteo_clases = df['Pclass'].value_counts().sort_index()
# conteo_clases.index -> los números de clase: [1, 2, 3]
# conteo_clases.values -> las cantidades correspondientes: ej. [216, 184, 491] (valores aproximados)

# --- Creación del gráfico de barras ---
# ax.bar() dibuja un gráfico de barras verticales.
# El resultado se guarda en la variable 'bar'. Este comando devuelve un contenedor de barras (BarContainer),
# que contiene objetos Rectangle para cada barra, permitiendo personalizaciones posteriores.
# Parámetros:
#   x = conteo_clases.index: las posiciones en el eje X (las clases 1, 2, 3).
#   height = conteo_clases.values: la altura de cada barra (las frecuencias).
#   color = ['gold', 'silver', '#cd7f32']: (opcional) asigna colores personalizados (oro, plata, bronce).
#   edgecolor = 'black': (opcional) añade un borde negro a las barras.
bar = ax.bar(conteo_clases.index, conteo_clases.values, color=['gold', 'silver', '#cd7f32'], edgecolor='black')

# --- Personalización del gráfico ---
ax.set_xlabel('Clase del pasajero')        # Etiqueta del eje X
ax.set_ylabel('Número de pasajeros')       # Etiqueta del eje Y
ax.set_title('Cantidad de pasajeros por clase en el Titanic')  # Título
ax.set_xticks([1, 2, 3])                   # Fuerza a que las marcas del eje X sean exactamente 1, 2, 3
ax.grid(axis='y', linestyle='--', alpha=0.7)  # Cuadrícula horizontal

plt.show()
```
