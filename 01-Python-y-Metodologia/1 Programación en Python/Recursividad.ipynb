# EJERCICIO 1

```python
def invierte_cadena(cadena):
    return cadena[::-1]

# Ejemplo de uso
print(invierte_cadena("123abc"))  # Salida: "cba321"
```

Slicing [::-1]:

El primer : indica que se tome toda la cadena.

El segundo : después del primero separa el paso (step).

-1 como paso indica que la cadena se recorra desde el último carácter hasta el primero, invertiendo el orden.

# EJERCICIO 2

```python
import math

def prueba_primo(n):
    # Los números menores o iguales a 1 no son primos
    if n <= 1:
        return False
    # El 2 es el único número par primo
    if n == 2:
        return True
    # Los números pares mayores que 2 no son primos
    if n % 2 == 0:
        return False
    # Verificar divisores impares desde 3 hasta la raíz cuadrada de n
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True

# Ejemplo de uso
print(prueba_primo(7))  # Salida: True
```

¿Para qué sirve?
El módulo math proporciona acceso a:

Funciones matemáticas comunes:
math.sqrt(x) - Calcula la raíz cuadrada

math.pow(x, y) - Calcula x elevado a y

math.sin(x), math.cos(x), math.tan(x) - Funciones trigonométricas

math.log(x) - Logaritmo natural

math.factorial(x) - Factorial de un número

Constantes matemáticas:
math.pi - Valor de π (aproximadamente 3.14159)

math.e - Valor de e (aproximadamente 2.71828)
