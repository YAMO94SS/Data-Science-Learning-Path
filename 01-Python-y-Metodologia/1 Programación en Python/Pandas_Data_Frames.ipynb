```python
# ============================================
# ACTIVIDAD: ANÁLISIS DE DATOS IRIS CON PANDAS
# ============================================

import pandas as pd
# Esta línea evita warnings innecesarios por encadenamiento de asignaciones
pd.options.mode.chained_assignment = None

# ------------------------------------------------------------
# 1. CARGAR EL DATAFRAME DESDE EL ARCHIVO CSV
# ------------------------------------------------------------
# Ruta completa del archivo (ajusta si es necesario)
ruta = r"C:\Users\Yamil\OneDrive\Escritorio\Escritorio\Yamil\EBAC (CIENCIA DE DATOS)\7 MÓDULO Pandas; Dataframes, lectura y exportación de archivos\7.4 Explora creación y visualización de Dataframes\iris.csv"
df = pd.read_csv(ruta)

# ------------------------------------------------------------
# PARTE 1
# ------------------------------------------------------------
# 2. FILTRAR REGISTROS CON ANCHO DE SÉPALO >= 3.5 Y LARGO > 5
#    Se usan paréntesis para cada condición y & para combinarlas
df_filtrado = df[(df['sepal.width'] >= 3.5) & (df['sepal.length'] > 5)]

# 3. CONTAR EL NÚMERO DE REGISTROS DEL DATAFRAME FILTRADO
registros = len(df_filtrado)

# 4. CALCULAR MEDIA, MEDIANA Y DESVIACIÓN ESTÁNDAR DEL ANCHO DEL PÉTALO POR CADA ESPECIE
#    Agrupamos por 'variety' y aplicamos las funciones de agregación
agregados = df_filtrado.groupby('variety')['petal.width'].agg(
    mean='mean',      # media
    median='median',  # mediana (percentil 50)
    std='std'         # desviación estándar
).reset_index()       # convertimos el índice 'variety' en columna

# ------------------------------------------------------------
# PARTE 2
# ------------------------------------------------------------
# 5. CALCULAR MEDIA Y DESVIACIÓN ESTÁNDAR DE TODA LA COLUMNA 'petal.width' (en el df_filtrado)
media_petal_width = df_filtrado['petal.width'].mean()
std_petal_width = df_filtrado['petal.width'].std()

# 6. CREAR NUEVA COLUMNA CON VALORES ESTANDARIZADOS (Z-SCORE)
#    Fórmula: (x - media) / desviación estándar
df_filtrado['petal.width.st'] = (df_filtrado['petal.width'] - media_petal_width) / std_petal_width

# 7. FILTRAR SOLO LAS COLUMNAS 'variety' Y LA NUEVA COLUMNA ESTANDARIZADA
df_final = df_filtrado[['variety', 'petal.width.st']].copy()  # .copy() evita problemas de vistas

# 8. RENOMBRAR LAS COLUMNAS CON LOS NOMBRES SOLICITADOS
df_final.rename(columns={'variety': 'Variedad', 'petal.width.st': 'Ancho Est. Pétalo'}, inplace=True)

# (Opcional) Mostrar los resultados para verificar
print("Registros filtrados:", registros)
print("\nAgregados por especie:\n", agregados)
print("\nPrimeras filas de df_final:\n", df_final.head())
```
