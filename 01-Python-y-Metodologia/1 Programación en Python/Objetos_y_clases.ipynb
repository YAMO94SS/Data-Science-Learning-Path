```python
class Conversion:
    def romano_entero(self, romano):
        # Valores de los números romanos
        valores = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }

        total = 0
        largo = len(romano)

        for i in range(largo):
            actual = valores[romano[i]]

            # Si hay un símbolo después y es mayor, restamos (ej. IV, XL, CM)
            if i + 1 < largo and actual < valores[romano[i + 1]]:
                total -= actual
            else:
                total += actual

        return total


# --- PRUEBA ---
# Crear objeto
numero = Conversion()

# Probar con los ejemplos del enunciado
print(numero.romano_entero('MMMCMLXXXVI'))  # 3986
print(numero.romano_entero('VIII'))         # 8
print(numero.romano_entero('XLIV'))         # 44
print(numero.romano_entero('CLXX'))         # 170
print(numero.romano_entero('MCDL'))         # 1450
```
