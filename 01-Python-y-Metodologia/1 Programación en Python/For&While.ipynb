En caso de que el valor proporcionado sea inválido (por ejemplo, que sea negativo o fraccionado), deberá devolver un mensaje de "Error de captura". Pruebe su código calculando los factoriales de 4, 5 y 6 respectivamente así como el de 7.5.

-El factorial de un número entero positivo n, representado por n!, es la multiplicación de n por sus antecesores mayores o iguales a 1. Por ejemplo: 4! = 4x3x2x1 = 24:

```python
cadena = input("Dame un entero positivo: ")

# Puedes usar la estructura "try/except" que aprendimos en el módulo 2.
# comience su código aquí:

factorial = None

try:
    # intentamos convertir a entero
    n = int(cadena)
     # Verificamos si es negativo
    if n < 0:
        print("Error de captura")
        else:
        # Calcular el factorial
        factorial = 1
        for i in range(1, n + 1):
            factorial *= i
        print(factorial)
        
except ValueError:
    # Si no se puede convertir a entero, mostrar error
    print("Error de captura")
```
